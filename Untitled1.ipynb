```python
#use of pandas
from sklearn.datasets import load_iris
import pandas as pd

df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
print(df.shape)
df.head()
```

```python
#iris.data contains only the feature values (like sepal length, petal width, etc.) for each sample.
#iris.target is a separate array that holds the numeric species code (0, 1, or 2) for each sample.
iris = load_iris()
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
df.head()
```

```python
import matplotlib.pyplot as plt
plt.hist(df['sepal length (cm)'], bins=20, color='skyblue', edgecolor='black')
plt.title('Sepal Length Distribution')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns
sns.scatterplot(x='petal length (cm)', y='petal width (cm)', hue='species', data=df)
plt.title('Petal Length vs Petal Width')
plt.show()
```
